```python
# 라이브러리 가져오기
import fitz, io, os, numpy as np, warnings
from paddleocr import PaddleOCR
from google.colab import files
from PIL import Image
from pykospacing import Spacing

# 경고 메시지 숨기기 (깔끔한 출력을 위해)
warnings.filterwarnings("ignore", category=DeprecationWarning)

# OCR 및 띄어쓰기 교정기 초기화 (한번만 로딩하면 됩니다)
print("필요한 AI 모델들을 준비 중입니다...")
ocr = PaddleOCR(use_angle_cls=True, lang='korean')
spacing = Spacing()
print("준비 완료!")

# --- 텍스트 처리 함수들 ---

def fix_ocr_spacing(ocr_raw_data, y_tol=1, x_col_gap=1):
    """OCR 결과의 라인/단 분할 및 띄어쓰기를 교정합니다."""
    if not ocr_raw_data or not ocr_raw_data[0]: return ""

    lines = {} # Y축 기준으로 텍스트 박스들을 묶습니다.
    for box, (text_content, _) in ocr_raw_data[0]:
        y, x_start, x_end = box[0][1], box[0][0], box[1][0]
        y_key = round(y / y_tol) * y_tol
        lines.setdefault(y_key, []).append({'x_s': x_start, 'x_e': x_end, 'text': text_content})

    fixed_lines = []
    for y_key in sorted(lines): # Y축 정렬 후 각 라인 처리
        boxes_on_line = sorted(lines[y_key], key=lambda i: i['x_s'])
        if not boxes_on_line: continue

        line_txt = boxes_on_line[0]['text']
        for i in range(1, len(boxes_on_line)):
            prev, curr = boxes_on_line[i-1], boxes_on_line[i]
            gap = curr['x_s'] - prev['x_e']
            if gap > x_col_gap: line_txt += '\t' # 단이 분리되면 탭
            elif gap > 0: line_txt += ' '       # 아니면 공백
            line_txt += curr['text']
        fixed_lines.append(spacing(line_txt)) # 한 줄 띄어쓰기 교정 후 추가

    return "\n".join(fixed_lines)

def fix_text_spacing(raw_text):
    """일반 텍스트(선택 가능 텍스트)의 줄 단위 띄어쓰기를 교정합니다."""
    if not raw_text.strip(): return ""

    # 줄 단위로 분리하여 각 줄에 띄어쓰기 적용 후 다시 합칩니다.
    processed_lines = [spacing(line).strip() for line in raw_text.split('\n') if line.strip()]
    return "\n".join(processed_lines)

# --- 메인 파일 처리 로직 ---

print("파일 업로드를 기다리고 있습니다...")
uploaded_files = files.upload() # 파일 업로드

for filename, content in uploaded_files.items():
    print(f"\n'{filename}' 파일 처리 시작...")
    file_ext = os.path.splitext(filename)[1].lower()

    if file_ext == '.pdf':
        doc = fitz.open(stream=content, filetype="pdf")
        pdf_s_text_output = "" # PDF 내 선택 가능 텍스트 결과
        pdf_o_text_output = "" # PDF 내 이미지 OCR 텍스트 결과

        for page_idx in range(len(doc)):
            page = doc[page_idx]

            # 1. PDF 선택 가능 텍스트 처리
            sel_text_raw = page.get_text()
            if sel_text_raw.strip():
                pdf_s_text_output += f"--- [페이지 {page_idx + 1}] 선택 가능 텍스트 ---\n{fix_text_spacing(sel_text_raw)}\n\n"

            # 2. PDF 내 이미지 텍스트 (OCR) 처리
            img_in_page = page.get_images(full=True)
            if img_in_page:
                pdf_o_text_output += f"--- [페이지 {page_idx + 1}] 이미지 텍스트 (OCR) ---\n"

            for img_idx, img_info in enumerate(img_in_page):
                img_data = doc.extract_image(img_info[0])
                try:
                    img = Image.open(io.BytesIO(img_data["image"])).convert('RGB')
                    ocr_result = ocr.ocr(np.array(img), cls=True)
                    pdf_o_text_output += f"[이미지 {img_idx + 1}]\n{fix_ocr_spacing(ocr_result)}\n"
                except Exception as e:
                    print(f"이미지 처리 오류 (페이지: {page_idx+1}, 이미지: {img_idx+1}): {e}")

        if pdf_s_text_output: print("\nPDF 선택 가능 텍스트 결과:\n", pdf_s_text_output)
        if pdf_o_text_output: print("\nPDF 이미지 텍스트 (OCR) 결과:\n", pdf_o_text_output)
        doc.close()

    elif file_ext in ['.jpg', '.jpeg', '.png', '.bmp', '.gif']:
        try:
            img = Image.open(io.BytesIO(content)).convert('RGB')
            ocr_result = ocr.ocr(np.array(img), cls=True)
            print("\n이미지 파일 텍스트 (OCR) 결과:\n", fix_ocr_spacing(ocr_result))
        except Exception as e:
            print(f"이미지 파일 처리 오류: {e}")

    else:
        print("지원되지 않는 파일 형식입니다. PDF 또는 이미지 파일을 업로드해주세요.")

print("\n모든 파일 처리가 완료되었습니다!")
```
